```python
nr, nc = 2, 2 
c = (0)
fig, axes = plt.subplots(nrows=nr, ncols=nc, figsize=(3*6.4,3*4.8))
for i in range(nr):
    for j in range(nc):
        axes[i, j].imshow(plt.imread(filenames[c]))
        axes[i, j].title.set_text(filenames[c])
        axes[i, j].axis('off')
        c += 1
fig.tight_layout()
plt.show() 
```
